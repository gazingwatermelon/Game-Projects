```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python

os.getcwd()
```

```python
movie_data = pd.read_csv('movies.csv')
movie_data
```

```python
#index, homepage, id, overview, original_title, release_date, overview, runtime, crew, vote_average, production_company, production_country, spoken_language and vote_count not needed to find order of recommendations
movie_data = movie_data.drop(['homepage','id','original_title','release_date', 'overview', 'index', 'runtime', 'vote_count', 'crew', 'vote_average', 'production_companies', 'production_countries', 'spoken_languages'], axis=1)
movie_data
```

```python
'''1|if a movie is entered as input then the features whose matching will determine order of recommendationa are:
        1.keywords
        2.genres
        3.original language
        4.cast
        5.director
        6.status
        7.tagline
'''
'''2|After the matching ordering must be done such that the movie with higher:
        1.Profit
        2.Popularity
   comes first.
   
   There there are three steps to find 
'''
#pre-processing genres, keywords and cast
genres = np.array(movie_data['genres'])
for i in range(len(genres)):
    if(type(genres[i])==float):
        genres[i]=[]
        continue
    genres[i]=genres[i].split()

keywords = np.array(movie_data['keywords'])
for i in range(len(keywords)):
    if(type(keywords[i])==float):
        keywords[i]=[]
        continue
    keywords[i]=keywords[i].split()

cast = np.array(movie_data['cast'])
for i in range(len(keywords)):
    if(type(cast[i])==float):
        cast[i]=[]
        continue
    cast[i] = cast[i].split()

title_components = np.array(movie_data['title'])
for i in range(len(keywords)):
    if(type(cast[i])==float):
        title_components[i]=[]
        continue
    title_components[i] = title_components[i].split()
```

```python
mod_data = pd.DataFrame()
mod_data['genres'] = genres
mod_data['keywords'] = keywords
mod_data['tagline'] = movie_data['tagline']
mod_data['original_language'] =  movie_data['original_language']
mod_data['status'] = movie_data['status']
mod_data['cast'] = cast
mod_data['director'] = movie_data['director']
mod_data['Profit'] = movie_data['revenue']-movie_data['budget']
mod_data['popularity'] = (movie_data['popularity']-movie_data['popularity'].min())/( movie_data['popularity'].max()-movie_data['popularity'].min())
mod_data['title'] = movie_data['title']

mod_data
```

```python
selection = input("Enter Movie Name ")
```

```python
genre_interest = mod_data[mod_data['title']==selection].iloc[0,0]
l = len(genres)
genre_score = np.zeros(l)
for i in range(l):
    count = 0
    genre_check = mod_data.iloc[i,0]
    for j in genre_check:
        if j in genre_interest:
            count+=1
    genre_score[i] = count/len(genre_interest)
print(genre_score)
```

```python
keywords_interest = mod_data[mod_data['title']==selection].iloc[0,1]
keywords_score = np.zeros(l)
for i in range(l):
    count = 0
    keywords_check = mod_data.iloc[i,1]
    for j in keywords_check:
        if j in keywords_interest:
            count+=1
    keywords_score[i] = count/len(keywords_interest)
print(keywords_score)
```

```python
original_language_interest = mod_data[mod_data['title']==selection].iloc[0,3]
original_language_score = np.zeros(l)
for i in range(l):
    if (original_language_interest == mod_data.iloc[i,3]):
        original_language_score[i] = 1
print(original_language_score)
```

```python
status_interest = mod_data[mod_data['title']==selection].iloc[0,4]
status_score = np.zeros(l)
for i in range(l):
    if (status_interest == mod_data.iloc[i,4]):
        status_score[i] = 1
print(status_score)
```

```python
cast_interest = mod_data[mod_data['title']==selection].iloc[0,5]
cast_score = np.zeros(l)
for i in range(l):
    count = 0
    cast_check = mod_data.iloc[i,5]
    for j in cast_check:
        if j in cast_interest:
            count+=1
    cast_score[i] = count/len(cast_interest)
print(cast_score)
```

```python
tagline_interest = mod_data[mod_data['title']==selection].iloc[0,2]
tagline_score = np.zeros(l)
for i in range(l):
    if (tagline_interest == mod_data.iloc[i,2]):
        tagline_score[i] = 1
        print(i)
print(tagline_score)
```

```python
director_interest = mod_data[mod_data['title']==selection].iloc[0,6]
director_score = np.zeros(l)
for i in range(l):
    if (director_interest == mod_data.iloc[i,6]):
        director_score[i] = 1
print(director_score)
```

```python
title_interest = mod_data[mod_data['title']==selection].iloc[0,9]
title_score = np.zeros(l)
for i in range(l):
    if(mod_data.iloc[i,9][-1].isdigit() and title_interest[-1].isdigit()):
        if (mod_data.iloc[i,9][:-1] == title_interest[:-1]):
            title_score[i] = 1
    elif(mod_data.iloc[i,9][-1].isdigit() and not title_interest[-1].isdigit()):
        if (mod_data.iloc[i,9][:-2] == title_interest):
            title_score[i] = 1
    elif(not mod_data.iloc[i,9][-1].isdigit() and title_interest[-1].isdigit()):
        if (mod_data.iloc[i,9] == title_interest[:-2]):
            title_score[i] = 1
print(title_score)
```

```python
mod_data_2 = pd.DataFrame()
mod_data_2['title'] = mod_data['title']
mod_data_2['tagline_score'] = tagline_score
mod_data_2['keywords_score'] = keywords_score
mod_data_2['genre_score'] = genre_score
mod_data_2['original_language_score'] = original_language_score
mod_data_2['status_score'] = status_score
mod_data_2['cast_score'] = cast_score
mod_data_2['director_score'] = director_score
mod_data_2['popularity'] = mod_data['popularity']
mod_data_2['success_of_movie'] = (mod_data['Profit']-mod_data['Profit'].min())/( mod_data['Profit'].max()-mod_data['Profit'].min())
mod_data_2['title_score'] = title_score
mod_data_2['overall_score'] = mod_data_2['keywords_score']*350 + mod_data_2['genre_score']*300+mod_data_2['director_score']*200+mod_data_2['cast_score']*250+mod_data_2['tagline_score']*10000+mod_data_2['original_language_score']*50+mod_data_2['status_score']*30+mod_data_2['popularity']*150+mod_data_2['success_of_movie']*100+mod_data_2['title_score']*1000
# tagline_score>title_score>keyword_score>genre_score>cast_score>director_score>popularity>success_of_movie>original_languge_score>status_score is the weightage order
mod_data_2
```
